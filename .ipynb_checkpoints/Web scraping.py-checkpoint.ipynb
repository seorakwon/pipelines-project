```python

from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd

def GetSoup(url):
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')
    return soup
```

```python
url = 'http://www.planecrashinfo.com/2019/2019.htm'

GetSoup(url)
```

```python
tags = soup.find_all('a')
tags
```

```python
for tag in tags:
    print(tag.get('href'))
```

```python
accident_table = soup.find_all('table')

accident_table
```

```python
rows = accident_table.find_all('tr')
accidents = []

for row in rows:
    cell = row.find_all('td')
    accident = {
        "Year": cell[0].find('a').text
    }
    accidents.append(accident)

print(accidents)

```
